```python
#Linear and Circular Convolution
import numpy as np
import math 
import scipy as sy
import matplotlib.pyplot as plt

x = [-10, -20, -30, -40] 
h = [20, -4, -5, 0]

n1 = list(range(0, len(x)))
n2 = list(range(0, len(h))) 
```

```python
#plotting input signals
plt.stem(n1, x)
plt.title("Input x[n]", fontsize= 20)
plt.xlabel('No. of samples n', fontsize= 18)
plt.ylabel('Amplitude x', fontsize= 18)
```

```python
plt.stem(n2, h)
plt.title("Input h[n]", fontsize= 20)
plt.xlabel('No. of samples n', fontsize= 18)
plt.ylabel('Amplitude x', fontsize= 18) 
```

```python
# Linear Convolution
N1 = len(x)
N2 = len(h)
N = N1+N2-1
y1 = np.zeros(N)
m = N-N1
n = N-N2
#Padding zeros to x and h to make their length to N
x =np.pad(x,(0,m),'constant')
h =np.pad(h,(0,n),'constant')

for n in range (N):
    for k in range (N):
        if n >= k:
           y1[n] = y1[n]+x[n-k]*h[k]
        
print('Linear convolution y =\n',y1)
plt.stem(y1)
plt.title("Linear Convolution", fontsize= 20)
plt.xlabel('No. of samples n', fontsize= 18)
plt.ylabel('Amplitude x', fontsize= 18)
plt.yticks(np.arange(min(y1), max(y1)+1, 100.0))
```

```python
#function for circular convolution
def circonvolution(x, h, n, m):  
    if(n>m):
        maxSize = n
    else:
        maxSize = m
        
    row_vec = [0]*maxSize
    col_vec = [0]*maxSize
    out = [0]*maxSize
    circular_shift_mat = [[0 for i in range(n)] for j in range(maxSize)]
        
    for i in range(maxSize): 
        if (i >= n): 
            row_vec[i] = 0; 
        else: 
            row_vec[i] = x[i]; 
            
    for i in range(maxSize): 
        if (i >= m): 
            col_vec[i] = 0; 
        else: 
            col_vec[i] = h[i]; 
            
    # Generating 2D matrix of 
    # circularly shifted elements 
    k = 0; 
    d = 0; 
  
    for i in range(maxSize): 
        curIndex = k - d; 
        for j in range(maxSize): 
            circular_shift_mat[j][i] = row_vec[curIndex % maxSize] 
            curIndex += 1
          
        k = maxSize; 
        d += 1; 
    
    # Computing result by matrix 
    # multiplication and printing results 
    for i in range(maxSize): 
        for j in range(maxSize): 
            out[i] += circular_shift_mat[i][j]*col_vec[j];       
    return out
        
ycir = circonvolution(x, h, N1, N2)
print('Circular convolution y =\n',ycir)
plt.stem(ycir)
plt.title("Circular Convolution", fontsize= 20)
plt.xlabel('No. of samples n', fontsize= 18)
plt.ylabel('Amplitude x', fontsize= 18) 
```
